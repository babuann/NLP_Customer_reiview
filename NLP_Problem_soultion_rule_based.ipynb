# Aspect Term Extraction

```python
# Rule-based Extraction: Define a list of aspect terms that are generally used in clothing reviews
predefined_aspects = ["fit",
    "comfort",
    "material/fabric",
    "quality",
    "style",
    "color",
    "pattern",
    "sizing",
    "length",
    "neckline",
    "sleeve style",
    "closure",
    "pockets",
    "collar type",
    "hemline",
    "stitching",
    "lining",
    "sheerness",
    "weight",
    "breathability",
    "stretchiness",
    "ease of care",
    "versatility"
]


# Create a new column to store the extracted aspects for each review
dress['Extracted_Aspects'] = None

# Loop through the cleaned review text to identify and store the aspect terms
for aspect in predefined_aspects:
    mask = dress['Cleaned_Review_Text'].str.contains(aspect)
    dress.loc[mask, 'Extracted_Aspects'] = dress.loc[mask, 'Extracted_Aspects'].apply(lambda x: x + [aspect] if x else [aspect])

# Show the first few rows containing the extracted aspects
dress[['Cleaned_Review_Text', 'Extracted_Aspects']].head(10)
```

```python
from collections import Counter
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag

# Tokenize the cleaned review text and perform POS tagging
all_reviews = ' '.join(dress['Cleaned_Review_Text'])
tokenized_reviews = word_tokenize(all_reviews)
pos_tags = pos_tag(tokenized_reviews)

# Extract all nouns
nouns = [word for word, pos in pos_tags if pos in ['NN', 'NNS']]

# Count the frequency of each noun
noun_freq = Counter(nouns)

# Show the 10 most common nouns, which could be potential aspect terms
common_nouns = noun_freq.most_common(10)
common_nouns
```

### classify the sentiment expressed towards each of these aspects.

```python
# list of positive and negative words for sentiment analysis.

positive_words = ['good', 'great', 'excellent', 'amazing', 'wonderful', 'love', 'like','elegant', 'stylish', 'chic', 'flattering', 'beautiful', 'classy', 'sophisticated', 'versatile', 'comfortable', 'well-made', 'high-quality', 'fashionable', 'gorgeous', 'unique', 'feminine', 'timeless', 'charming', 'lovely', 'luxurious', 'graceful', 'stunning', 'amazing', 'trendy', 'impeccable', 'exquisite', 'alluring', 'perfect', 'fabulous', 'fantastic', 'superb', 'impressive', 'splendid', 'outstanding', 'exceptional', 'dazzling', 'splendid', 'captivating', 'dashing', 'excellent', 'delightful', 'awe-inspiring', 'striking', 'flawless', 'adorable', 'uplifting', 'brilliant', 'radiant']
negative_words = [
    "uncomfortable",
    "itchy",
    "tight",
    "loose",
    "poorly-made",
    "cheap",
    "scratchy",
    "unflattering",
    "wrinkles easily",
    "see-through",
    "stiff",
    "too short",
    "too long",
    "ill-fitting",
    "dated",
    "frumpy",
    "unstylish",
    "sloppy",
    "low-quality",
    "flimsy",
    "overpriced",
    "disappointing",
    "lacks structure",
    "not durable",
    "poor stitching",
    "color fades",
    "awkward fit",
    "inconvenient",
    "impractical",
    "limited mobility",
    "bad",
    "poor",
    "terrible",
    "awful",
    "hate",
    "dislike"
]


# Initialize a column to store the sentiment scores
dress['Aspect_Sentiment'] = None

# Function to score sentiment based on the presence of positive and negative words
def score_sentiment(sentence):
    pos_count = sum(1 for word in sentence.split() if word in positive_words)
    neg_count = sum(1 for word in sentence.split() if word in negative_words)
    return pos_count - neg_count

# Loop through each review to perform sentiment analysis for each aspect
for index, row in dress.iterrows():
    review_text = row['Cleaned_Review_Text']
    aspects = row['Extracted_Aspects']
    if aspects:
        aspect_sentiment = {}
        for aspect in aspects:
            # Find sentences that contain the aspect
            sentences = [sent for sent in sent_tokenize(review_text) if aspect in sent]
            # Score sentiment for each sentence and average
            if sentences:
                sentiment_score = sum(score_sentiment(sent) for sent in sentences) / len(sentences)
                aspect_sentiment[aspect] = sentiment_score
        dress.at[index, 'Aspect_Sentiment'] = aspect_sentiment


dress[['Cleaned_Review_Text', 'Extracted_Aspects', 'Aspect_Sentiment']].head(10)
```

#### In the 'Aspect_Sentiment' column, you can see a dictionary that holds the sentiment score for each aspect mentioned in the corresponding review.

A positive score indicates a positive sentiment.
A negative score indicates a negative sentiment.
A score of zero suggests a neutral sentiment.

### visualizing the data 

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Initialize a dictionary to store the summed sentiment scores for each aspect and the count of mentions
aspect_sentiment_sum = defaultdict(int)
aspect_count = defaultdict(int)

# Loop through the DataFrame to sum the sentiment scores for each aspect
for index, row in dress.iterrows():
    aspect_sentiments = row['Aspect_Sentiment']
    if aspect_sentiments:
        for aspect, sentiment in aspect_sentiments.items():
            aspect_sentiment_sum[aspect] += sentiment
            aspect_count[aspect] += 1

# Calculate the average sentiment for each aspect
aspect_sentiment_avg = {aspect: aspect_sentiment_sum[aspect] / aspect_count[aspect] for aspect in aspect_sentiment_sum}

# Sort aspects by average sentiment
sorted_aspects = sorted(aspect_sentiment_avg.items(), key=lambda x: x[1])

# Data for plotting
aspects = [x[0] for x in sorted_aspects]
avg_sentiments = [x[1] for x in sorted_aspects]

# Create the plot
plt.figure(figsize=(12, 6))
plt.barh(aspects, avg_sentiments, color=['red' if x < 0 else 'green' for x in avg_sentiments])
plt.xlabel('Average Sentiment')
plt.ylabel('Aspects')
plt.title('Average Sentiment for Each Aspect in clothing Id 1078')
plt.show()
```
